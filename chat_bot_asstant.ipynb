# How to use this Chat Assistant Notebook

- Make sure you have a valid Groq API key and set it in the environment variable as shown in Cell 4.
- Run all cells in order for a working chat assistant CLI.
- Type your message at the prompt. Type 'bye' or 'exit' to end the chat.
- The chat history is preserved in the session and sent to the LLM for context.
- You can modify the system prompt in Cell 5 to change the assistant's behavior.

# langchain v1 + Groq + create_agent + simple cli chat with chat history


```python
import os
```

```python
os.environ["GROQ_API_KEY"] = "<YOUR_GROQ_API_KEY>"  # Set your Groq API key securely, e.g., via environment variable
```

```python
from langchain.agents import create_agent

# configure the LLM

agent = create_agent(
    model="groq:llama-3.1-8b-instant",
    tools=[],
    system_prompt="You are a helpful chat assistant. Be clear, concise, and polite. Understand the user’s intent and respond directly. Stay professional and safe."
)
result = agent.invoke(
    {
        "messages": [{"role": "user", "content": "Explain Machine Learning in short"}]
    }
)
```

```python
print(result)
```

```python
print(result["messages"][1].content)
```

```python
# chat history list
chat_history = []


print("🤖 Chat Assistant ready! Type 'bye' or 'exit' to stop.\n")

# simple chat loop
while True:
  user_input = input("You: ".strip())
  print("\n")
  if user_input.lower() in ["bye", "exit"]:
    
    print("Assistant: Goodbye 👋")
    break

  messages = chat_history + [{"role": "user", "content": user_input}]
  result = agent.invoke({"messages": messages})

  # extract reply
  try:
    reply = result["messages"][-1].content
  except Exception as e:
    reply = str(e)
  print(f"{user_input}\n")
  print(f"Asistant: {reply}\n")
  print("-"*60)

  # update chat history
  chat_history.append({"role": "user", "content": user_input})
  chat_history.append({"role": "assistant", "content": reply})
```
